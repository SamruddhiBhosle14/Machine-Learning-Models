```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
```

```python
dataset.info()
```

```python
dataset.head(5)
```

```python
# Classify dataset for input and output

X = dataset[['Age' , 'EstimatedSalary']]
y = dataset['Purchased']
```

```python
# same standardization 

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
```

```python
X = sc.fit_transform(X)
```

```python
# split into training and test data

from sklearn.model_selection import train_test_split

Xtrain , Xtest , ytrain , ytest = train_test_split(X,y,test_size = 0.25,random_state = 0)
```

```python
# train a knn model

from sklearn.neighbors import KNeighborsClassifier
```

```python
classifier = KNeighborsClassifier(n_neighbors = 5)
```

```python
classifier.fit(Xtrain ,ytrain)
```

```python
pred = classifier.predict(Xtest)
```

```python
err = ytest - pred

err.value_counts()
```

```python
# Evaluate using a Confusion matrix

from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(ytest , pred)

print(cm)
```

```python
# Visualizing The KNN

from matplotlib.colors import ListedColormap

sns.set_style('whitegrid')
plt.figure(figsize = (20,10))

Xset , yset = Xtrain , ytrain

X1 , X2 = np.meshgrid(np.arange(Xset[:,0].min() - 1 , Xset[:,0].max() + 1, 0.01) ,
                      np.arange(Xset[:,1].min() - 1  , Xset[:,1].max() + 1, 0.01))
plt.contourf(X1,X2,classifier.predict(np.array([X1.ravel() , X2.ravel()]).T).reshape(X1.shape) , 
             alpha = 0.30 , cm = ListedColormap(('red' , 'green')))

plt.xlim(X1.min(),X1.max())
plt.ylim(X2.min(),X2.max())

for i , j in enumerate(np.unique(yset)):
        plt.scatter(Xset[yset == j , 0] , Xset[yset == j , 1] , c = ListedColormap(('red','green'))(i) , label = j)
        
plt.title('KNN for Training Data')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.legend()
```

We can see that the decision boundary here is not a linear line

so we can classify the data more efficiently 

IF we increase the number of neighbors the we may end up overfitting the model

```python
from matplotlib.colors import ListedColormap

sns.set_style('whitegrid')
plt.figure(figsize = (20,10))

Xset , yset = Xtest , ytest

X1 , X2 = np.meshgrid(np.arange(Xset[:,0].min() - 1 , Xset[:,0].max() + 1, 0.01) ,
                      np.arange(Xset[:,1].min() - 1  , Xset[:,1].max() + 1, 0.01))
plt.contourf(X1,X2,classifier.predict(np.array([X1.ravel() , X2.ravel()]).T).reshape(X1.shape) , 
             alpha = 0.30 , cm = ListedColormap(('red' , 'green')))

plt.xlim(X1.min(),X1.max())
plt.ylim(X2.min(),X2.max())

for i , j in enumerate(np.unique(yset)):
        plt.scatter(Xset[yset == j , 0] , Xset[yset == j , 1] , c = ListedColormap(('red','green'))(i) , label = j)
        
plt.title('KNN for Test Data')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.legend()
```
