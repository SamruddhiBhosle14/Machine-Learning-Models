# The goal is to detect if the hired employee is bluffing about his/her previous salary or not , which according to him/her  is  160K

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
```

```python
dataset.info()
```

```python
dataset.head()
```

```python
sns.lineplot(x = 'Level' , y = 'Salary' , data = dataset)
```

As we don't have much data available we will not split the data

So we avoid importing of model_selection

```python
X = pd.DataFrame(data = dataset['Level'])
y = pd.DataFrame(data = dataset['Salary'])
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
```

```python
lin_reg.fit(X,y)
```

The linear model is trained but it does not have any polynomial features

To Add the features we will create a new set of independent variable

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
# intially we tried with degree 2 and then with 3 , 4
# as we increase the degree we fit the model more precisely...
# but remember not to overfit your model

poly_reg = PolynomialFeatures(degree = 4)

# this will create a polynomial feature object that wil transfrom and fit our orignal independent data 
```

```python
X_poly = poly_reg.fit_transform(X)
```

```python
poly_reg.fit(X_poly , y)
```

Again we train a linear model using a new linear model...

The first model is built to compare the two regression models and none of the are dependent on each other!

```python
lin_reg_1 = LinearRegression()
```

```python
lin_reg_1.fit(X_poly , y)
```

```python
# Visualizing the linear regression model lin_reg

sns.set_style('whitegrid')
plt.figure(figsize = (12,7))

plt.scatter(X , y , color = 'red')
plt.plot(X , lin_reg.predict(X) , color = 'blue')
plt.title('Truth Or Bluff(Linear Regression)')
plt.xlabel('Levels')
plt.ylabel('Salary')
```

As we can see this is a very bad fit 

So we go for polynomial regression

```python
#Visualizing the Polynomial Regression 

plt.figure(figsize = (12,7))
plt.scatter(X , y , color = 'red')
plt.plot(X , lin_reg_1.predict(poly_reg.fit_transform(X)) , color = 'blue')

plt.title('Truth Or Bluff(Polynomial Regression for degree 2)')
plt.xlabel('Levels')
plt.ylabel('Salary')
```

```python
#Visualizing the Polynomial Regression 

plt.figure(figsize = (12,7))
plt.scatter(X , y , color = 'red')
plt.plot(X , lin_reg_1.predict(poly_reg.fit_transform(X)) , color = 'blue')

plt.title('Truth Or Bluff(Polynomial Regression for degree 4)')
plt.xlabel('Levels')
plt.ylabel('Salary')
```

As we can see the degree 4 model is a better fit to the data.

But when we look closely at level = 2 , 7 ,8 ,9 we can see that the curve is not smooth

This is due to lack of data

but for now we can compensate by using following steps

```python
X_grid = np.arange(min(X['Level']) ,max(X['Level']) , 0.1)
```

```python
np.size(X_grid)
```

```python
X_grid = pd.DataFrame(data = X_grid)
```

```python
#Visualizing the Polynomial Regression wiith X_grid

plt.figure(figsize = (12,7))
plt.scatter(X , y , color = 'red')
plt.plot(X_grid , lin_reg_1.predict(poly_reg.fit_transform(X_grid)) , color = 'blue')

plt.title('Truth Or Bluff(Polynomial Regression for degree 4 with X_grid)')
plt.xlabel('Levels')
plt.ylabel('Salary')
```

This is more of a smooth curve..
