```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset.head(5)
```

```python
dataset.info()
```

```python
plt.figure(figsize=(18,12))

sns.scatterplot(x = dataset['Annual Income (k$)'] , y= dataset['Spending Score (1-100)'] ,color = 'green')
```

```python
X = dataset[['Annual Income (k$)' , 'Spending Score (1-100)']].values
```

```python
# No standarization is required as the column values lie in the same range

from sklearn.cluster import KMeans
```

```python
# Picking up the right number of clusters using the elbow method by calculating the wcss

wcss = []

for i in range(1,11):
    kmeans = KMeans(n_clusters = i , init = 'k-means++' , n_init = 10 , max_iter = 300 , random_state = 0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    
plt.figure(figsize=(16,10))
plt.plot(range(1,11) , wcss)
plt.title('Elbow Method')
plt.xlabel('Range')
plt.ylabel('WCSS')

# init is used to specify the initialization of the cluster centroids , as random intialization can be inefficient 
```

```python
# as we get that at 5 clusters there is the elbow , do we choose 5 clusters

kmeans = KMeans(n_clusters = 5 , init = 'k-means++' , max_iter = 300 , n_init = 10 , random_state = 0)
pred_clusters = kmeans.fit_predict(X)
```

```python
pred_clusters
```

```python
# the final coordinates of the centroids is obtained by cluster_centers_

kmeans.cluster_centers_
```

```python
#Visuzlize the clusters
sns.set_style(style='darkgrid')
plt.figure(figsize=(20,16))

plt.scatter(X[pred_clusters == 0 , 0],X[pred_clusters == 0 , 1] , s = 100 , c = 'red', label = 'Cluster1')
plt.scatter(X[pred_clusters == 1 , 0],X[pred_clusters == 1 , 1] , s = 100 , c = 'blue', label = 'Cluster2')
plt.scatter(X[pred_clusters == 2 , 0],X[pred_clusters == 2 , 1] , s = 100 , c = 'green', label = 'Cluster3')
plt.scatter(X[pred_clusters == 3 , 0],X[pred_clusters == 3 , 1] , s = 100 , c = 'magenta', label = 'Cluster4')
plt.scatter(X[pred_clusters == 4 , 0],X[pred_clusters == 4 , 1] , s = 100 , c = 'orange', label = 'Cluster5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1] , s = 300 , c = 'black', label = 'Centroid')

plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Clusters')
plt.legend()
```
